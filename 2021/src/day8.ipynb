<p>The problem is that the signals which control the segments have been mixed up on each display. Worse, the wire/segment connections are mixed up separately for each four-digit display! (All of the digits <em>within</em> a display use the same connections, though.)</p>
<p>So, you might know that only signal wires <code>b</code> and <code>g</code> are turned on, but that doesn't mean <em>segments</em> <code>b</code> and <code>g</code> are turned on: the only digit that uses two segments is <code>1</code>, so it must mean segments <code>c</code> and <code>f</code> are meant to be on. <p>For each display, you watch the changing signals for a while, make a note of <em>all ten unique signal patterns</em> you see, and then write down a single <em>four digit output value</em> (your puzzle input).
```python
from more_itertools import flatten


example_small = "acedgfb cdfbe gcdfa fbcad dab cefabd cdfgeb eafb cagedb ab | cdfeb fcadb cdfeb cdbaf"
example_large = """
be cfbegad cbdgef fgaecd cgeb fdcge agebfd fecdb fabcd edb | fdgacbe cefdb cefbgd gcbe
edbfga begcd cbg gc gcadebf fbgde acbgfd abcde gfcbed gfec | fcgedb cgb dgebacf gc
fgaebd cg bdaec gdafb agbcfd gdcbef bgcad gfac gcb cdgabef | cg cg fdcagb cbg
fbegcd cbd adcefb dageb afcb bc aefdc ecdab fgdeca fcdbega | efabcd cedba gadfec cb
aecbfdg fbg gf bafeg dbefa fcge gcbea fcaegb dgceab fcbdga | gecf egdcabf bgf bfgea
fgeab ca afcebg bdacfeg cfaedg gcfdb baec bfadeg bafgc acf | gebdcfa ecba ca fadegcb
dbcfg fgd bdegcaf fgec aegbdf ecdfab fbedc dacgb gdcebf gf | cefg dcbef fcge gbcadfe
bdfegc cbegaf gecbf dfcage bdacg ed bedf ced adcbefg gebcd | ed bcgafe cdgba cbgef
egadfb cdbfeg cegd fecab cgb gbdefca cg fgcdab egfdb bfceg | gbdfcae bgc cg cgb
gcafb gcf dcaebfg ecagb gf abcdeg gaef cafbge fdbac fegbdc | fgae cfgab fg bagce
"""


def count_digits1_4_7_8(s: str) -> int:
    output_values = flatten(
        re.split(r"\s+\|\s+", l)[1].split(" ") for l in s.strip().splitlines()
    )
    # length -> digit: 2 -> 1 , 4 -> 4, 3 -> 7, 7 -> 8
    return sum(1 for v in output_values if len(v) in (2, 3, 4, 7))


assert count_digits1_4_7_8(example_large) == 26
```

```python
from more_itertools import one


def create_number(patterns, values) -> int:
    length_to_patterns = defaultdict(list)

    for p in patterns.split(" "):
        length_to_patterns[len(p)].append(p)

    identiefied = {
        1: one(length_to_patterns[2]),
        4: one(length_to_patterns[4]),
        7: one(length_to_patterns[3]),
        8: one(length_to_patterns[7]),
    }

    identiefied[3] = one(
        p for p in length_to_patterns[5] if all(s in p for s in identiefied[1])
    )
    identiefied[6] = one(
        p for p in length_to_patterns[6] if not all(s in p for s in identiefied[1])
    )
    identiefied[2] = one(
        p
        for p in length_to_patterns[5]
        if p != identiefied[3] and sum(1 for s in p if s in identiefied[6]) == 4
    )
    identiefied[5] = one(
        p for p in length_to_patterns[5] if p != identiefied[2] and p != identiefied[3]
    )
    identiefied[9] = one(
        p
        for p in length_to_patterns[6]
        if p != identiefied[6] and sum(1 for s in p if s in identiefied[4]) == 4
    )

    identiefied[0] = one(
        p for p in length_to_patterns[6] if p != identiefied[6] and p != identiefied[9]
    )

    value_to_digit = {tuple(sorted(v)): d for d, v in identiefied.items()}

    return int(
        "".join(str(value_to_digit[tuple(sorted(v))]) for v in values.split(" "))
    )


def map_wires_segments(s: str):
    return sum(create_number(*re.split(r"\s+\|\s+", l)) for l in s.strip().splitlines())


assert map_wires_segments(example_large) == 61229
```

```python
# Nice solution from https://github.com/viliampucik/adventofcode/blob/master/2021/08.py
# Changed to match to play with match case
part1 = part2 = 0

for line in puzzle.splitlines():
    signals, output = line.strip().split(" | ")

    d = {l: set(s) for s in signals.split() if (l := len(s)) in (2, 4)}

    n = ""
    for o in output.split():
        match len(o):
            case 2:
                n += "1"
                part1 += 1
            case 4:
                n += "4"
                part1 += 1
            case 3:
                n += "7"
                part1 += 1
            case 7:
                n += "8"
                part1 += 1
            case 5:
                s = set(o)
                if len(s & d[2]) == 2:
                    n += "3"
                elif len(s & d[4]) == 2:
                    n += "2"
                else:
                    n += "5"
            case 6:  # l == 6
                s = set(o)
                if len(s & d[2]) == 1:
                    n += "6"
                elif len(s & d[4]) == 4:
                    n += "9"
                else:
                    n += "0"
    part2 += int(n)

print(part1)
print(part2)
```
